```python
hrpdata1 = hrpdata.drop(["Date",'Unnamed: 14'],axis=1)
hrpdata1.isna().sum()
```

```python
def weightshrp (df1) :
    port = rp.HCPortfolio(returns=df1)
    model='HRP'
    codependence = 'pearson'
    rm = 'MV'
    rf = 0
    linkage = 'single'
    max_k = 10
    leaf_order = True
    w = port.optimization(model=model,
                          codependence=codependence,
                          rm=rm,
                          rf=rf,
                          linkage=linkage,
                          max_k=max_k,
                          leaf_order=leaf_order)
    return(w)
```

```python
weightshrp(hrpdata1)
```
