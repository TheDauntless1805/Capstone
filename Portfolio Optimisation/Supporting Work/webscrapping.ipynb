```python
# Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Create an URL object
url = 'https://www.screener.in/company/WONDERLA/'
# Create object page
page = requests.get(url)
```

```python
page
```

```python
# parser-lxml = Change html to Python friendly format
# Obtain page's information
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
div_bs4 = soup.find_all('span',attrs={'class':'number'})
div_bs4
```

# CAGR Table

```python
tables = soup.find_all("table", attrs={"class": "ranges-table"})
```

```python
yr = ['10 Years:','5 Years:','3 Years:']
for table in tables:
    th = table.find('th')
    print(th.text)
    for i in yr:
        td = table.find('td',text= i)
        td = td.findNext('td')
        print(i,td.text)
```
