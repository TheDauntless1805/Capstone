### Applying Monte Carlo simulations to Portfolio Management

"We are going to use a Monte Carlo simulation to simulate and plot thousands of portfolios and pick the ones with the highest returns by plotting returns against standard deviation. These portfolios are called by the author “The Efficient Frontier”.

- The intuition is to use a Monte Carlo simulation to assign random weights to each of these tickers in our portfolio (the sum of which will add to one) and plot the returns of each the portfolio against standard deviation.

### creating random 10000 portfolios of random combinations


```python
portfolioReturns = []  ## empty list for portfolio returns
portfolioRisk = []  ## empty list for portfolio Risk
stockWeights = []   ## ## empty list for asset weights
```

```python
num_assets = len(df1.columns) 
num_portfolios = 10000
```

```python
cov
```

```python
for portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)

    stockWeights.append(weights)
    returns = np.dot(weights, annualReturns) # Returns are the product of individual expected returns of asset and its
                                        # weights
    portfolioReturns.append(returns)
    var = cov.mul(weights, axis=0).mul(weights, axis=1).sum().sum()# Portfolio variance
    sd = np.sqrt(var) # paily standard deviation
    ann_sd = sd*np.sqrt (250) # Annual standard deviation - volatility
    portfolioRisk.append(ann_sd)
```

```python
weights.shape
```

```python
data = {'Returns':portfolioReturns,'Volatility':portfolioRisk}
len(data['Volatility'])
```

```python
## Selecting minimum variance Portfolio

for counter, symbol in enumerate(df1.columns.to_list()):
    data[symbol+'_weight'] = [w[counter] for w in stockWeights]
portfolio = pd.DataFrame(data)
```

```python
rf = 0.06122
portfolio['Sharpe_Ratio'] = (portfolio['Returns'] - rf)/portfolio["Volatility"]
```

```python
def gglaso_function(df,alpha = None):
    
    ##Calling Glasso algorithm
#     covariance_matrix /= covariance_matrix.std(axis=0)

    # Creating Model 
    model = GraphicalLasso(alpha = alpha).fit(df)
    
    ## Returning the Precision Matrix
    return (model.precision_)
```

```python
def min_variance_portfolio_function(portfolio):
    min_vol_port = portfolio.iloc[portfolio['Volatility'].idxmin()]
    return min_vol_port

def optimised_portfolio_function(portfolio):
    optimum_port = portfolio.iloc[portfolio['Sharpe_Ratio'].idxmax()]
    return optimum_port
    
```

```python
def min_optimised_port_function(portfolio):
    result = pd.DataFrame(columns=portfolio.columns).T
    
    # getting minimum portfolio
    min_vol_port = portfolio.iloc[portfolio['Volatility'].idxmin()]

    # getting the optimum portfolio
    optimum_port = portfolio.iloc[portfolio['Sharpe_Ratio'].idxmax()]
    
    # adding portfolios to the result df
    result['Minimum Variance Portfolio'] = min_vol_port
    result['Optimum Portfolio'] = optimum_port
    
    
    return result

```

```python
def graph(portfolio,result):
    # Plotting Optimum Portfolio
    plt.subplots(figsize=(20,10))
    plt.scatter(x=portfolio['Volatility'],y=portfolio['Returns'],marker = 'o',s=10)
    plt.scatter(x=result.loc["Volatility"][0],y=result.loc["Returns"][0],color='red',marker="*",s=500)
    plt.scatter(x=result.loc["Volatility"][1],y=result.loc["Returns"][1],color='green',marker="*",s=500)
    plt.show()
```

```python
def portfolios_function(df,annualReturns, cov,num_portfolios=10000,rf=0.06122):
    portfolioReturns = []  ## empty list for portfolio returns
    portfolioRisk = []  ## empty list for portfolio Risk
    stockWeights = []   ## ## empty list for asset weights

    num_assets = len(df.columns) 

    for portfolio in range(num_portfolios):
        weights = np.random.random(num_assets)
        weights /= np.sum(weights)

        stockWeights.append(weights)
        returns = np.dot(weights, annualReturns) # Returns are the product of individual expected returns of asset and its
                                            # weights
        portfolioReturns.append(returns)
        var = cov.mul(weights, axis=0).mul(weights, axis=1).sum().sum()# Portfolio variance
        sd = np.sqrt(var) # paily standard deviation
        ann_sd = sd*np.sqrt (250) # Annual standard deviation - volatility
        portfolioRisk.append(ann_sd)

    data = {'Returns':portfolioReturns,'Volatility':portfolioRisk}

    ## Selecting minimum variance Portfolio

    for counter, symbol in enumerate(df.columns.to_list()):
        data[symbol+'_weight'] = [w[counter] for w in stockWeights]
    portfolio = pd.DataFrame(data)
    portfolio['Sharpe_Ratio'] = (portfolio['Returns'] - rf)/portfolio["Volatility"]

    return portfolio
```

```python
def portfolio_function(df,alpha,n_portfolio):
    
    # Calculating means of the stocks
    annualReturns = df.mean()*252
    
    # Calculating orignal covariance matrix
    cov = df.cov()*252
    
    # Generating new covariance matrix by using gglasso   
    precision_matrix = gglaso_function(df,alpha)
    
    cov = pd.DataFrame(np.linalg.inv(precision_matrix) * 252)
    
#     return cov
    
    # Generating multiple portfolio
    portfolios = portfolios_function(df,annualReturns,cov)
    # Getting minimum variance portfolio
    result = min_optimised_port_function(portfolios)
    print(result)
    # Getting Maximum Sharpe Portfolio
    
    graph(portfolios,result)
```
